### **Tech Challenge - Modelo Preditivo de Diagnóstico de Obesidade**

**Objetivo:** Desenvolver um modelo de Machine Learning capaz de prever o nível de obesidade ou sobrepeso de um indivíduo com base em suas características demográficas, físicas e hábitos comportamentais para auxiliar a tomada de decisão da equipe médica.
**Meta Técnica:** Assertividade (Acurácia) mínima acima de 75% no conjunto de teste e deploy da aplicação preditiva em produção.

Dados extraídos do prontuário/questionário hospitalar: `Obesity.csv`

#### **Informações importantes sobre a base de dados**

* **Gender:** Gênero do paciente.
* **Age:** Idade.
* **Height:** Altura em metros.
* **Weight:** Peso em quilogramas (kgs).
* **family_history:** Histórico familiar (algum membro sofreu ou sofre de excesso de peso?).
* **FAVC:** Consumo frequente de alimentos altamente calóricos.
* **FCVC:** Frequência de consumo de vegetais nas refeições.
* **NCP:** Número de refeições principais feitas diariamente.
* **CAEC:** Consumo de alimentos entre as refeições.
* **SMOKE:** Paciente possui hábito de fumar?
* **CH2O:** Quantidade de água consumida diariamente.
* **SCC:** Monitoramento diário das calorias ingeridas.
* **FAF:** Frequência de prática de atividade física.
* **TUE:** Tempo diário de uso de dispositivos tecnológicos (celular, computador, videogame, etc.).
* **CALC:** Frequência de consumo de álcool.
* **MTRANS:** Meio de transporte mais utilizado no cotidiano.
* **Obesity (Variável Alvo):** Classificação do nível de obesidade/sobrepeso do paciente.

A base de dados conta com **2.111 registros** sem valores nulos, divididos de forma equilibrada entre as classes, o que permitirá o treinamento de algoritmos multiclasse e o desenvolvimento de um painel analítico com foco em saúde preventiva.

# **7. Estratégia de Engenharia de Atributos**

Para evitar o **Vazemento de dados** (Data Leakage), realizamos primeiro a separação do dataset entre variáveis preditoras (X) e variável alvo (y), seguida da divisão em conjuntos de Treino e Teste.

As transformações categóricas e numéricas serão estruturadas através de um **ColumnTransformer** para serem integradas diretamente ao Pipeline de Machine Learning.

```python
# Separação das Variáveis e Split

from sklearn.model_selection import train_test_split

# 1. Separação em Preditoras (X) e Alvo (y)
X = df_tratado.drop(columns=['nivel_obesidade'])
y = df_tratado['nivel_obesidade']

# 2. Divisão em Treino (80%) e Teste (20%) com estratificação para manter as classes balanceadas
X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print(f"Dados de Treino: {X_treino.shape[0]} registros")
print(f"Dados de Teste: {X_teste.shape[0]} registros")
```

```python
# Definição da Estrutura de Feature Engineering

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder

# Identificando as colunas por tipo de tratamento
colunas_numericas = ['idade', 'altura', 'peso', 'frequencia_consumo_vegetais',
                     'numero_refeicoes_principais', 'consumo_agua_diario',
                     'frequencia_atividade_fisica', 'tempo_dispositivos_tecnologicos']

colunas_categoricas_nominais = ['genero', 'historico_familiar_sobrepeso',
                                'consome_alimentos_alto_calorico', 'fumante',
                                'monitora_calorias_diarias', 'meio_transporte']

# Variáveis categóricas ordinais (que possuem uma ordem lógica de intensidade)
colunas_categoricas_ordinais = ['consome_entre_refeicoes', 'frequencia_consumo_alcool']
ordem_categorias = [
    ['Nunca', 'Às vezes', 'Frequentemente', 'Sempre'], # Ordem para consome_entre_refeicoes
    ['Nunca', 'Às vezes', 'Frequentemente', 'Sempre']  # Ordem para frequencia_consumo_alcool
]

# Criando o transformador que será injetado no Pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), colunas_numericas),
        ('cat_nominal', OneHotEncoder(drop='first', handle_unknown='ignore'), colunas_categoricas_nominais),
        ('cat_ordinal', OrdinalEncoder(categories=ordem_categorias), colunas_categoricas_ordinais)
    ]
)

print("Estrutura de Feature Engineering criada com sucesso!")
```

# **8. Treinamento e Comparação de Modelos**

Nesta etapa, estabelecemos um modelo baseline utilizando **Regressão Logística** e o comparamos com algoritmos mais complexos (**Random Forest** e **XGBoost**).
Para garantir a integridade dos testes, todos os modelos serão validados no conjunto de teste utilizando as métricas de **Acurácia** (Meta > 75%) e **F1-Score Macro** (para avaliar o desempenho balanceado entre as 7 classes de diagnóstico).

```python
# Instalação do XGBoost e Importação das Métricas

from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder

# Importante: O XGBoost exige que as classes (y) sejam numéricas (0 a 6)
# Vamos codificar a variável alvo apenas para o treino/teste dos modelos
label_encoder = LabelEncoder()
y_treino_encoded = label_encoder.fit_transform(y_treino)
y_teste_encoded = label_encoder.transform(y_teste)

# Definindo o dicionário com os modelos que vamos testar
modelos = {
    'Regressão Logística (Baseline)': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    'XGBoost': XGBClassifier(random_state=42, eval_metric='mlogloss')
}

# Lista para armazenar o dicionário de resultados
resultados_lista = []

print("Modelos e métricas configurados")
```

## **8.1 Cenário 1: Treinamento dos Modelos com Peso e Altura**

Para avaliar a capacidade do modelo em antecipar o risco de obesidade utilizando apenas o histórico familiar e os hábitos cotidianos do paciente, removemos utilizando as variáveis preditoras diretas de massa física (`peso` e `altura`) e reavaliamos os algoritmos.

```python
# ==============================================================================
# Treinamento e Avaliação
# ==============================================================================

# Loop dos modelos
for nome, modelo in modelos.items():
    # Criando o Pipeline completo: Preprocessamento + Algoritmo
    pipeline_completo = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', modelo)
    ])

    # Treinando o pipeline (fit aprende os tratamentos e o modelo APENAS no treino)
    pipeline_completo.fit(X_treino, y_treino_encoded)

    # Predizendo no conjunto de teste
    y_pred = pipeline_completo.predict(X_teste)

    # Calculando as métricas
    acuracia = accuracy_score(y_teste_encoded, y_pred)
    f1 = f1_score(y_teste_encoded, y_pred, average='macro')

    # Salvando os resultados
    resultados_lista.append({
        'Modelo': nome,
        'Acurácia (Teste)': acuracia,
        'F1-Score (Macro)': f1
    })

# Convertendo a lista de resultados em um DataFrame para exibição elegante
df_resultados = pd.DataFrame(resultados_lista)

print("=== Tabela Comparativa de Performance ===")
display(df_resultados.sort_values(by='Acurácia (Teste)', ascending=False))
```

```python
# ==============================================================================
# Treinamento e Avaliação dos Modelos (Formato de Texto - Padrão Fase 2)
# ==============================================================================
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import LabelEncoder

# Importação dos algoritmos solicitados
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# 1. O XGBoost exige que as classes sejam numéricas (0 a 6).
# Vamos codificar a variável alvo de treino e teste.
label_encoder = LabelEncoder()
y_treino_encoded = label_encoder.fit_transform(y_treino)
y_teste_encoded = label_encoder.transform(y_teste)

print("=== INICIANDO AVALIAÇÃO DOS MODELOS ===\n")

# --- MODELO 1: Regressão Logística (Baseline) ---
pipe_logistic = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', LogisticRegression(max_iter=1000, random_state=42))])
pipe_logistic.fit(X_treino, y_treino_encoded)
y_pred_logistic = pipe_logistic.predict(X_teste)
print(f"Acurácia - Regressão Logística (Base Line): {accuracy_score(y_teste_encoded, y_pred_logistic):.2%}")

# --- MODELO 2: KNN ---
pipe_knn = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', KNeighborsClassifier())])
pipe_knn.fit(X_treino, y_treino_encoded)
y_pred_knn = pipe_knn.predict(X_teste)
print(f"Acurácia - KNN: {accuracy_score(y_teste_encoded, y_pred_knn):.2%}")

# --- MODELO 3: SVM ---
pipe_svm = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', SVC(random_state=42))])
pipe_svm.fit(X_treino, y_treino_encoded)
y_pred_svm = pipe_svm.predict(X_teste)
print(f"Acurácia - SVM: {accuracy_score(y_teste_encoded, y_pred_svm):.2%}")

# --- MODELO 4: Random Forest ---
pipe_rf = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', RandomForestClassifier(random_state=42))])
pipe_rf.fit(X_treino, y_treino_encoded)
y_pred_rf = pipe_rf.predict(X_teste)
print(f"Acurácia - Random Forest: {accuracy_score(y_teste_encoded, y_pred_rf):.2%}")

# --- MODELO 5: XGBoost ---
pipe_xgb = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', XGBClassifier(random_state=42, eval_metric='mlogloss'))])
pipe_xgb.fit(X_treino, y_treino_encoded)
y_pred_xgb = pipe_xgb.predict(X_teste)
print(f"Acurácia - XGBoost: {accuracy_score(y_teste_encoded, y_pred_xgb):.2%}")

print("\n" + "="*50 + "\n Evaluation concluída!")
```

Para avaliar a capacidade do modelo em antecipar o risco de obesidade utilizando apenas o histórico familiar e os hábitos cotidianos do paciente, removemos as variáveis preditoras diretas de massa física (`peso` e `altura`) e reavaliamos os algoritmos.

```python
# ==============================================================================
# Preparação dos Dados e Pipeline sem Peso e Altura
# ==============================================================================
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.model_selection import train_test_split

# 1. Removendo peso e altura das variáveis preditoras
X_comportamental = df_tratado.drop(columns=['nivel_obesidade', 'peso', 'altura'])
y_comportamental = df_tratado['nivel_obesidade']

# 2. Divisão em Treino e Teste (Mantendo a mesma semente 42 para comparação justa)
X_treino_c, X_teste_c, y_treino_c, y_teste_c = train_test_split(
    X_comportamental, y_comportamental,
    test_size=0.20,
    random_state=42,
    stratify=y_comportamental
)

# Encodando a variável alvo
le_c = LabelEncoder()
y_treino_c_encoded = le_c.fit_transform(y_treino_c)
y_teste_c_encoded = le_c.transform(y_teste_c)

# 3. Atualizando o Pré-processador (Removendo 'peso' e 'altura' da lista numérica)
colunas_numericas_c = ['idade', 'frequencia_consumo_vegetais', 'numero_refeicoes_principais',
                       'consumo_agua_diario', 'frequencia_atividade_fisica', 'tempo_dispositivos_tecnologicos']

preprocessor_c = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), colunas_numericas_c),
        ('cat_nominal', OneHotEncoder(drop='first', handle_unknown='ignore'), colunas_categoricas_nominais), # Mantém as mesmas nominais
        ('cat_ordinal', OrdinalEncoder(categories=ordem_categorias), colunas_categoricas_ordinais)       # Mantém as mesmas ordinais
    ]
)

print("Novo conjunto comportamental estruturado com sucesso!")
```

```python
# ==============================================================================
# 8.2 Torneio de Modelos Sem Peso e Altura (Exibição em Texto - Padrão Fase 2)
# ==============================================================================
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

print("=== INICIANDO AVALIAÇÃO DOS MODELOS (SEM PESO E ALTURA) ===\n")

# --- MODELO 1: Regressão Logística ---
pipe_logistic_c = Pipeline(steps=[('preprocessor', preprocessor_c), ('classifier', LogisticRegression(max_iter=1000, random_state=42))])
pipe_logistic_c.fit(X_treino_c, y_treino_c_encoded)
y_pred_logistic_c = pipe_logistic_c.predict(X_teste_c)
print(f"Acurácia - Regressão Logística (Base Line): {accuracy_score(y_teste_c_encoded, y_pred_logistic_c):.2%}")

# --- MODELO 2: KNN ---
pipe_knn_c = Pipeline(steps=[('preprocessor', preprocessor_c), ('classifier', KNeighborsClassifier())])
pipe_knn_c.fit(X_treino_c, y_treino_c_encoded)
y_pred_knn_c = pipe_knn_c.predict(X_teste_c)
print(f"Acurácia - KNN: {accuracy_score(y_teste_c_encoded, y_pred_knn_c):.2%}")

# --- MODELO 3: SVM ---
pipe_svm_c = Pipeline(steps=[('preprocessor', preprocessor_c), ('classifier', SVC(random_state=42))])
pipe_svm_c.fit(X_treino_c, y_treino_c_encoded)
y_pred_svm_c = pipe_svm_c.predict(X_teste_c)
print(f"Acurácia - SVM: {accuracy_score(y_teste_c_encoded, y_pred_svm_c):.2%}")

# --- MODELO 4: Random Forest ---
pipe_rf_c = Pipeline(steps=[('preprocessor', preprocessor_c), ('classifier', RandomForestClassifier(random_state=42))])
pipe_rf_c.fit(X_treino_c, y_treino_c_encoded)
y_pred_rf_c = pipe_rf_c.predict(X_teste_c)
print(f"Acurácia - Random Forest: {accuracy_score(y_teste_c_encoded, y_pred_rf_c):.2%}")

# --- MODELO 5: XGBoost ---
pipe_xgb_c = Pipeline(steps=[('preprocessor', preprocessor_c), ('classifier', XGBClassifier(random_state=42, eval_metric='mlogloss'))])
pipe_xgb_c.fit(X_treino_c, y_treino_c_encoded)
y_pred_xgb_c = pipe_xgb_c.predict(X_teste_c)
print(f"Acurácia - XGBoost: {accuracy_score(y_teste_c_encoded, y_pred_xgb_c):.2%}")

print("\n" + "="*50 + "\n Nova avaliação concluída!")
```
